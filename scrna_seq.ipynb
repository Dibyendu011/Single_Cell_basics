#### Selecting top 2000 variable genes (highest variation across cells) :
##### We bin all genes based on expression levels (all bins have a different set of genes with no common genes in two bins). Variability is measured using z-score. 
##### z_score = (gene_variance - bin_mean_variance) / bin_std_variance

##### gene_variance -> The variance of a gene's expression across all cells

##### Bin statistics:
##### bin_mean_variance = 50 -> mean(variance_gene1, variance_gene2, ..., variance_geneN) where there are N genes in tht bean, variance is measured across all cells
##### bin_std_variance = 20

##### Individual genes:
##### Gene A: variance=100 → z_score = (100-50)/20 = 2.5
##### Gene B: variance=120 → z_score = (120-50)/20 = 3.5  

##### These 2000 genes with the highest z score across all bins are selected & rest of the genes can be  eliminated across all samples using "subset=True"

```python
sc.pp.highly_variable_genes(adata, n_top_genes = 2000, subset=True, flavor="seurat_v3")
adata
```

```python
### start from here
import os
os.chdir("/home/ibab/Desktop/DIBYENDU/s3/RA/scrna/GSE171524_RAW")
print(os.getcwd())
import scanpy as sc
import scvi
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ribo_genes_df = pd.read_csv('KEGG_RIBOSOME.v2025.1.Hs.tsv', sep='\t', header=None, names=['key', 'value'])
# Extract the gene symbols from the GENE_SYMBOLS row
gene_symbols_row = ribo_genes_df[ribo_genes_df['key'] == 'GENE_SYMBOLS']
ribo_gene_symbols = gene_symbols_row['value'].iloc[0].split(',')
np.array(ribo_gene_symbols)
```

```python
def pp(csv_path):
    testdf = pd.read_csv(csv_path, compression='gzip', index_col=0)
    adata = sc.AnnData(testdf).T
    sc.pp.filter_genes(adata, min_cells = 10)
    sc.pp.highly_variable_genes(adata, n_top_genes = 2000, subset = True, flavor = 'seurat_v3')
    warnings.filterwarnings("ignore", category=UserWarning, message="Can't initialize NVML")
    scvi.model.SCVI.setup_anndata(adata)
    vae = scvi.model.SCVI(adata)
    vae.train()
    solo = scvi.external.SOLO.from_scvi_model(vae)
    solo.train()
    df = solo.predict()
    df['prediction'] = solo.predict(soft = False)
    df.index = df.index.map(lambda x: x[:-2])
    df['dif'] = df.doublet - df.singlet
    doublets = df[(df.prediction == 'doublet') & (df.dif > 1)]
    
    adata = sc.read_csv(csv_path).T
    adata.obs['Sample'] = csv_path.split('_')[1]
    
    adata.obs['doublet'] = adata.obs.index.isin(doublets.index)
    adata = adata[~adata.obs.doublet]
    
    sc.pp.filter_cells(adata, min_genes=200)
    adata.var['mt'] = adata.var_names.str.startswith('mt-')
    adata.var['ribo'] = adata.var_names.isin(ribo_gene_symbols)
    sc.pp.calculate_qc_metrics(adata, qc_vars=['mt', 'ribo'], percent_top=None, log1p=False, inplace=True)
    upper_lim = np.quantile(adata.obs.n_genes_by_counts.values, .98)
    adata = adata[adata.obs.n_genes_by_counts < upper_lim]
    adata = adata[adata.obs.pct_counts_mt < 20]
    adata = adata[adata.obs.pct_counts_ribo < 2]

    return adata
```
